```python
def find_highest_freq_word_len(string):
    keep_track = {} 
    words = string.split()
    for word in words:
        keep_track[word] = keep_track.get(word,0)+1 
    max_freq = max(keep_track.values())
    
    max_freq_word = [word for word , freq in keep_track.items() if freq == max_freq]
    
    print (f"From the given string we can note that the most frequent words are {max_freq_word}  and its corresponding length is {len(max_freq_word[0])}")
    
```

```python
string = "write write write all the number from from from 1 to 100"
find_highest_freq_word_len(string)
```

```python
def is_valid(string):
    keep = {}
    for letter in string:
        keep[letter] = keep.get(letter,0)+1
        
    keep_freq = {}
    for count in keep.values():
        keep_freq[count] = keep.get(count,0)+1
        
    if len(keep_freq) == 1:
        return "Yes"
    else:
        return "No"
        
```

```python
is_valid("aaabbbcc")
```

```python
import pandas as pd
import urllib.request, json
```

```python
with urllib.request.urlopen('https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json') as url:
    data = json.load(url)
pokemon_data = []
for pokemon in data['pokemon']:    # converting the data into structured format
    processed_pokedata = {
        "id": pokemon["id"],
        "num": pokemon["num"],
        "name": pokemon["name"],
        "img": pokemon["img"],
        "type": pokemon["type"],
        "height": pokemon["height"],
        "weight": pokemon["weight"],
        "candy": pokemon.get("candy", ""),
        "candy_count": pokemon.get("candy_count", 0),
        "egg": pokemon.get("egg", ""),
        "spawn_chance": pokemon.get("spawn_chance", 0.0),
        "avg_spawns": pokemon.get("avg_spawns", 0),
        "spawn_time": pokemon.get("spawn_time", ""),
        "multipliers":pokemon.get("multipliers",[]),
        "weakness": pokemon.get("weaknesses", []),
        "next_evolution": [evolution["name"] for evolution in pokemon.get("next_evolution", [])],
        "prev_evolution": [evolution["name"] for evolution in pokemon.get("prev_evolution", [])]
    }
    pokemon_data.append(processed_pokedata)

poke_df = pd.DataFrame(pokemon_data) #converting processed data into pandas dataframe

poke_df.to_excel("pokemon_data.xlsx", index=False) # Exporting the dataframe to an excel file

```

```python
import csv
```

```python
with urllib.request.urlopen("https://data.nasa.gov/resource/y77d-th95.json") as url:
    data_meteorite = json.load(url)

    columns = [
        "id",
    "name",
    "nametype",
    "recclass",
    "mass",
    "fall",
    "year",
    "reclat",
    "reclong",
    "geolocation",
    ":@computed_region_cbhk_fwbd",
    ":@computed_region_nnqa_25f4"
    ]

# Create a CSV file and write the data
with open("meteorite_data.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=columns)

    # Write the header row
    writer.writeheader()

    # Write each meteorite data as a row in the CSV file
    for meteorite in data_meteorite:
        writer.writerow(meteorite)
```

```python
import requests

url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"
response = requests.get(url)
data_shows = response.json()

# Extract the desired data attributes
show_id = data_shows["id"]
show_url = data_shows["url"]
show_name = data_shows["name"]

episodes = data_shows["_embedded"]["episodes"]
extracted_data = []
for episode in episodes:
    season = episode["season"]
    number = episode["number"]
    episode_type = episode["type"]
    airdate = episode["airdate"]
    airtime = episode["airtime"]
    runtime = episode["runtime"]
    average_rating = episode["rating"]["average"]
    summary = episode["summary"].strip("<p>").strip("</p>")
    medium_image = episode["image"]["medium"]
    original_image = episode["image"]["original"]

    extracted_data.append({
        "id": show_id,
        "url": show_url,
        "name": show_name,
        "season": season,
        "number": number,
        "type": episode_type,
        "airdate": airdate,
        "airtime": airtime,
        "runtime": runtime,
        "average_rating": average_rating,
        "summary": summary,
        "medium_image": medium_image,
        "original_image": original_image
    })

# Print the extracted data
for episode_data in extracted_data:
    for key, value in episode_data.items():
        print(f"{key}: {value}")
    print()
```

```python
#Get all Pokemons that have no multipliers at all
no_multipliers = poke_data[poke_data['multipliers'].isnull()]
no_multipliers
```

```python
#Get all Pokemons that do not have more than 2 evolutions

#converting the string to list
def string_to_list(string):
    # Remove the surrounding square brackets and single quotes
    string = string.strip("[]").replace("'", "")
    
    # Split the remaining string by commas
    result_list = string.split(", ")
    
    return result_list

evo1 = [] #empty list to append each processed list 
for i in poke_data['next_evolution']:
    evo1.append(string_to_list(i))
poke_data['next_evolution_'] = evo1

#repeating the same with previous_evolution column
evo2=[]
for i in poke_data['prev_evolution']:
    evo2.append(string_to_list(i))
poke_data['prev_evolution_'] = evo2
```

```python
#so now that we have processed both the columns now we need to combine both into one column that is total+evolutions than we can filter out the pokemons that has mpore than 2 evolutions
poke_data['total_evolution'] = poke_data['next_evolution_']+poke_data['prev_evolution_']
poke_data
```

```python
#now we have a simgle colum  with total no. of evolutions.
no_more_than_2_evolutions = poke_data[poke_data['total_evolution'].str.count(',')<2]
```
